```python
import pandas as pd

df = pd.read_csv(
    r'C:\Users\kalas\OneDrive\Desktop\desktopcopy\coding\ML_Projects\House_Prices_Model\data\data\train.csv'
)
```

```python
y = df['SalePrice']
df = df.drop(columns=['SalePrice'])
```

```python
df = df.drop(columns=['Id'])

drop_cols = ['Alley','PoolQC','Fence','MiscFeature','FireplaceQu','MasVnrType']
df = df.drop(columns=drop_cols, errors='ignore')
```

```python
num_cols = df.select_dtypes(include=['int64','float64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

cat_cols = df.select_dtypes(include='object').columns
df[cat_cols] = df[cat_cols].fillna("None")
```

```python
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder()
df[cat_cols] = encoder.fit_transform(df[cat_cols])
```

```python
print(df.isnull().sum().sum())  # must be 0
print(df.shape)                # (1460, ~74)
print(y.shape)                 # (1460,)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    df, y, test_size=0.2, random_state=42
)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

rf = RandomForestRegressor(
    n_estimators=300,
    max_depth=15,
    min_samples_leaf=5,
    random_state=42,
    n_jobs=-1
)

rf.fit(X_train, y_train)

preds = rf.predict(X_val)
mae = mean_absolute_error(y_val, preds)

print("MAE:", mae)
```
